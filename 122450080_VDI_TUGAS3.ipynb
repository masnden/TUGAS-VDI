```python
# Let's first inspect the content of the file to understand its structure and the type of data it contains.
import pandas as pd

# Load the CSV file
file_path = '/content/student-mat.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Reload the data using the correct delimiter
data = pd.read_csv(file_path, delimiter=';')

# Display the first few rows of the dataset to verify the correction
data.head()
```

```python
import matplotlib.pyplot as plt

# Set up data for visualization
g3_data = data['G3']

# Good Visualization 1: Bar plot for distribution of final grades (G3)
plt.figure(figsize=(10, 6))
plt.hist(g3_data, bins=range(0, 21), color='skyblue', edgecolor='black')
plt.title('Distribution of Final Grades (G3)', fontsize=16)
plt.xlabel('Final Grade (G3)', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

Visualisasi di atas adalah Good Visualization dalam bentuk histogram yang menampilkan distribusi nilai akhir siswa (G3). Berikut alasannya berdasarkan empat prinsip visualisasi:

Kejelasan: Histogram ini dengan jelas menunjukkan bagaimana nilai akhir siswa didistribusikan, dengan kategori nilai yang jelas.

Kesederhanaan: Penggunaan histogram sederhana dan tidak menambahkan elemen yang tidak perlu, sehingga mudah dipahami.

Akuntabilitas: Data diambil langsung dari dataset asli, jadi representasi ini akurat.

Keterbacaan: Teks, sumbu, dan label skala mudah dibaca.

```python
# Bad Visualization 1: 3D Pie Chart for distribution of final grades (G3)
# For simplicity, we'll group the grades into categories for the pie chart.

# Group the grades into categories
g3_categories = pd.cut(g3_data, bins=[0, 5, 10, 15, 20], labels=['0-5', '6-10', '11-15', '16-20'])

# Count the number of students in each category
g3_category_counts = g3_categories.value_counts()

# Pie chart (3D style)
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
wedges, texts, autotexts = ax.pie(g3_category_counts, autopct="%1.1f%%", startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])

# Add title
plt.title('Distribution of Final Grades (G3) - 3D Pie Chart', fontsize=14)

# Show the plot
plt.show()
```

1. Tujuan Visualisasi:

Memahami Distribusi Nilai: Tujuan utama dari visualisasi ini adalah untuk memberikan gambaran yang jelas tentang bagaimana nilai akhir siswa (G3) didistribusikan, membantu pengguna melihat pola umum (misalnya, mayoritas siswa memperoleh nilai di rentang berapa).

Mengidentifikasi Kelompok Nilai Rendah atau Tinggi: Membantu mengidentifikasi kelompok siswa yang memiliki nilai rendah (0-5) dan siswa dengan nilai tinggi (16-20).
